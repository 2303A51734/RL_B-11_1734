```python
import numpy as np
import matplotlib.pyplot as plt

# GridWorld setup
GRID_SIZE = 5
GOAL = (4, 4)
PIT = (2, 2)
REWARD_GOAL = 10
REWARD_PIT = -10
REWARD_STEP = -1
GAMMA = 0.9
THRESHOLD = 1e-4  # Convergence threshold

# Possible actions
ACTIONS = ['R', 'D']  # Right, Down

def get_next_state(state, action):
    """Get next state given current state and action."""
    x, y = state
    if state == GOAL or state == PIT:
        return state  # Terminal states
    if action == 'R' and y < GRID_SIZE - 1:
        return (x, y + 1)
    if action == 'D' and x < GRID_SIZE - 1:
        return (x + 1, y)
    return state  # If action not possible, stay in place

def get_reward(state, next_state):
    """Get reward for transitioning to next_state."""
    if next_state == GOAL:
        return REWARD_GOAL
    if next_state == PIT:
        return REWARD_PIT
    if next_state == state:
        return 0  # No move, no penalty
    return REWARD_STEP

def policy(state):
    """Fixed policy: try Right, else Down."""
    if state == GOAL or state == PIT:
        return None  # No action in terminal states
    x, y = state
    if y < GRID_SIZE - 1:
        return 'R'
    return 'D'

def compute_value_function(gamma=GAMMA):
    """Compute value function using iterative policy evaluation."""
    V = np.zeros((GRID_SIZE, GRID_SIZE))
    while True:
        delta = 0
        for x in range(GRID_SIZE):
            for y in range(GRID_SIZE):
                state = (x, y)
                if state == GOAL or state == PIT:
                    continue  # Terminal states have fixed value 0
                v = V[x, y]
                action = policy(state)
                next_state = get_next_state(state, action)
                reward = get_reward(state, next_state)
                V[x, y] = reward + gamma * V[next_state]
                delta = max(delta, abs(v - V[x, y]))
        if delta < THRESHOLD:
            break
    return V

def visualize_value_function(V, title="Value Function"):
    """Visualize the value function as a heatmap."""
    plt.figure(figsize=(8, 6))
    plt.imshow(V, cmap='viridis', interpolation='nearest')
    plt.colorbar(label='Value')
    plt.title(title)
    plt.xticks(np.arange(GRID_SIZE))
    plt.yticks(np.arange(GRID_SIZE))
    for x in range(GRID_SIZE):
        for y in range(GRID_SIZE):
            plt.text(y, x, f'{V[x, y]:.2f}', ha='center', va='center', color='white')
    plt.grid(True)
    plt.show()

def analyze_gamma_values():
    """Analyze the effect of different gamma values."""
    gamma_values = [0.5, 0.9, 0.99]
    value_functions = {}
    for gamma in gamma_values:
        V = compute_value_function(gamma)
        value_functions[gamma] = V
        visualize_value_function(V, title=f"Value Function (Î³ = {gamma})")
    return value_functions

# Run the computation and visualization
V = compute_value_function()
print("Value Function (Î³ = 0.9):")
print(V)
visualize_value_function(V, "Value Function (Î³ = 0.9)")
value_functions = analyze_gamma_values()

# Analysis of gamma's effect
print("\nAnalysis of Discount Factor (Î³) Impact:")
for gamma, V in value_functions.items():
    print(f"\nÎ³ = {gamma}:")
    print(f"Value at start (0,0): {V[0,0]:.2f}")
    print(f"Value near goal (4,3): {V[4,3]:.2f}")
    print(f"Value near pit (2,1): {V[2,1]:.2f}")
```
